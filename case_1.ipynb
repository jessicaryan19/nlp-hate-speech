```python
import string
import random
import pickle
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords, gutenberg, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist
from nltk.tag import pos_tag
```

```python
import nltk
nltk.download('gutenberg')
```

```python
data = pd.read_csv('Assets/labeled_data.csv')

eng_stopwords = set(stopwords.words('english'))
porter = PorterStemmer()
wnl= WordNetLemmatizer()
data.head()
```

```python
data.dtypes
```

```python
list_words = []
for tweet in data['tweet']:
    if tweet:
        tweet = str(tweet)
        list_words.extend(word_tokenize(tweet))


list_words = [word.lower() for word in list_words if word not in eng_stopwords]
list_words = [word.lower() for word in list_words if word not in string.punctuation]
list_words = [word.lower() for word in list_words if word.isalpha()]
list_words = [word.lower() for word in list_words if word not in 'br']
list_words = [porter.stem(word) for word in list_words]
list_words = [wnl.lemmatize(word) for word in list_words]
print(list_words)
```

```python
tagged = pos_tag(list_words[:10])
print(tagged)
```

```python
# ner = ne_chunk(tagged)
# ner.draw()
```

```python
list_words = [w for w in list_words if w != "nan"]
```

```python
fd = FreqDist(list_words)
for word, freq in fd.most_common():
    print(word, freq)
```

```python
word = list_words[0]
synsets = wordnet.synsets(word)

for synset in synsets:
    print(f"{synset}: {synset.definition()}")
    for lemma in synset.lemmas():
        print(f"Synonim: {lemma.name()}")
        for antonym in lemma.antonyms():
            print(f"Antonym: {antonym.name()}")
```

```python
labeled_review = []
for tweet, label in zip(data['tweet'], data['class']):
    if label == 0.0 :
        labeled_review.append((str(tweet), 'Hate speech'))
    elif label == 1.0:   
        labeled_review.append((str(tweet), 'Offensive Language'))
    else:
        labeled_review.append((str(tweet), 'Neither'))

print(labeled_review)
```

```python
dataset = []
for tweet, sentiment in labeled_review:
    dict = {}
    words = [word.lower() for word in word_tokenize(tweet)]
    words = [word.lower() for word in words if word not in eng_stopwords]
    words = [word.lower() for word in words if word not in string.punctuation]
    words = [word.lower() for word in words if word.isalpha()]
    words = [word.lower() for word in words if word not in 'br']
    for feature in list_words:
        dict[feature] = feature in words
    dataset.append((dict, sentiment))
```

```python
from nltk.classify import NaiveBayesClassifier, accuracy
iter = 0
acc = 0

while acc < 0.8 and iter < 100:
    random.shuffle(dataset)
    counter = int(len(dataset) * 0.7)
    train = dataset[:counter]
    test = dataset[counter:]

    iter += 1
    classifier = NaiveBayesClassifier.train(train)
    acc = accuracy(classifier, test)
    
print(f"Accuracy: {acc}, Iteration: {iter}") 
```

```python
from nltk.classify import SklearnClassifier, accuracy
from sklearn.svm import LinearSVC
iter = 0
acc = 0

while acc < 0.8 and iter < 100:
    random.shuffle(dataset)
    counter = int(len(dataset) * 0.7)
    train = dataset[:counter]
    test = dataset[counter:]

    iter += 1
    classifier = SklearnClassifier(LinearSVC()).train(labeled_featuresets=train)
    acc = accuracy(classifier, test)
    
print(f"Accuracy: {acc}, Iteration: {iter}")

file = open('model/naive.pickle', 'wb')
classifier = pickle.dump(classifier, file)
file.close()
```
